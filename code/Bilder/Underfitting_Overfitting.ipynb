```
#code zum Erstellen von
from matplotlib import pyplot as plt
import numpy as np
import random
from google.colab import files
```

```
#--funktion erstellen--#
x  = np.arange(1, 50)
y1 = np.exp(-x/5) + .1
y2 = np.exp(-x/7) + .2
y3 = np.exp(-x/3) + .05
y4 = np.exp(-x/6) + .1

#--plotten--#
#underfitting
underfitting = plt.figure()
plt.title('Schematische Darstellung von Underfitting')
plt.xlabel('Epochen')
plt.ylabel('Fehlerfunktion')
plt.plot(x,y1, label='Train Loss')
plt.plot(x,y2, label='Validation Loss')
plt.ylim(0, 1)
plt.xlim(0,50)
#overfitting
overfitting = plt.figure()
plt.title('Schematische Darstellung von Overfitting')
plt.xlabel('Epochen')
plt.ylabel('Fehlerfunktion')
plt.plot(x,y1, label='Train Loss')
plt.plot(x,y3, label='Validation Loss')
plt.ylim(0, 1)
plt.xlim(0,50)
#optimal
opt = plt.figure()
plt.title('Schematische Darstellung von optimalem Fitting')
plt.xlabel('Epochen')
plt.ylabel('Fehlerfunktion')
plt.plot(x,y1, label='Train Loss')
plt.plot(x,y4, label='Validation Loss')
plt.ylim(0, 1)
plt.xlim(0,50)

#--anzeigen--#
underfitting.legend(bbox_to_anchor=(0.85, .85), loc='upper right', ncol=1)
underfitting.show()
overfitting.legend(bbox_to_anchor=(0.85, .85), loc='upper right', ncol=1)
overfitting.show()
opt.legend(bbox_to_anchor=(0.85, .85), loc='upper right', ncol=1)
opt.show()

#--download aus colab--#
underfitting.savefig('underfitting.png')
files.download('underfitting.png')
overfitting.savefig('overfitting.png')
#files.download('overfitting.png')
opt.savefig('optimal.png')
#files.download('optimal.png')
```
